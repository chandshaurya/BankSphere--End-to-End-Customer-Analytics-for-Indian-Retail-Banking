```python
# outlier detection

# IQR

def detect_outliers_iqr(df, columns):
    outlier_summary = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_summary[col] = {
            "Q1": Q1,
            "Q3": Q3,
            "IQR": IQR,
            "Lower Bound": lower_bound,
            "Upper Bound": upper_bound,
            "Outlier Count": outliers.shape[0],
            "Outlier %": (outliers.shape[0] / df.shape[0]) * 100
        }
    return pd.DataFrame(outlier_summary).T


# column analysis
columns_to_check = {
    "Customers": (cust, ["Age", "AnnualIncome"]),
    "Loans": (loan, ["loan_amount", "InterestRate"]),
    "Transactions": (trans, ["amount"]),
    "Profitability": (profit, ["RevenueGenerated", "NetProfitMargin"])
}

outlier_reports = {}

for name, (df, cols) in columns_to_check.items():
    report = detect_outliers_iqr(df, cols)
    outlier_reports[name] = report
    print(f"\n=== Outlier Analysis for {name} ===")
    print(report)

# Visualize all selected columns in subplots
for name, (df, cols) in columns_to_check.items():
    fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(6 * len(cols), 4))
    
    if len(cols) == 1:
        axes = [axes]  
    
    for ax, col in zip(axes, cols):
        sns.boxplot(x=df[col], ax=ax)
        ax.set_title(f"{name} - {col}", fontsize=12)
    
    plt.suptitle(f"Outlier Boxplots for {name}", fontsize=14, y=1.05)
    plt.tight_layout()
    plt.show()


```

```python
### CHURN Table


# Get the last transaction date per customer
last_txn = trans.groupby("customer_id")['date'].max().reset_index()
last_txn.rename(columns={'date': 'last_txn_date'}, inplace=True)

# Merge with customers
cust_churn = pd.merge(cust, last_txn, on="customer_id", how="left")

snapshot_date = trans['date'].max()  # e.g., latest date in data

# Add churn column
cust_churn['days_since_last_txn'] = (snapshot_date - cust_churn['last_txn_date']).dt.days
cust_churn['churn_status'] = cust_churn['days_since_last_txn'].apply(
    lambda x: "Churned" if x > 180 else "Active"
)


Churn_table = cust_churn[["customer_id", "CustomerName", "last_txn_date", "days_since_last_txn", "churn_status"]]
print(Churn_table.head())
```

```python
## Upsell Candidates by Segment

avg_gold_revenue = df_2[df_2['CustomerSegment'] == "Gold"]['RevenueGenerated'].mean()
print("Average Gold Revenue:", avg_gold_revenue)

# 2 Total revenue per customer
customer_revenue = df_2.groupby(['customer_id','CustomerSegment'], as_index=False)['RevenueGenerated'].sum()

#  Filter Silver/Regular customers whose revenue > avg_gold_revenue
upsell_candidates = customer_revenue[
    ((customer_revenue['CustomerSegment'].isin(["Silver","Regular"])) &
     (customer_revenue['RevenueGenerated'] > avg_gold_revenue))
]

#  Count
upsell_count = upsell_candidates.shape[0]

print("Upsell Candidates Count:", upsell_count)

upsell_by_segment = upsell_candidates.groupby("CustomerSegment")['customer_id'].count().reset_index()
upsell_by_segment.rename(columns={'customer_id':'Upsell_Count'}, inplace=True)

plt.figure(figsize=(6,4))
sns.barplot(x="CustomerSegment", y="Upsell_Count", data=upsell_by_segment,hue="CustomerSegment",legend=False,  palette="Greens")

plt.title("Upsell Candidates by Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Upsell Candidates")
plt.show()
```
